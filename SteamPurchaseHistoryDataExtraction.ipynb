```python
import pandas as pd

url = "https://raw.githubusercontent.com/YavuzCanAtalay/DSA210_TermProject/refs/heads/main/DataFile/SteamPurchaseHistoryRAW.csv"
df = pd.read_csv(url)

# Function to process the 'Total' column and handle mixed currencies
def process_total(total_value):
    if pd.isna(total_value):
        return 0  # Handle missing values as 0
    if 'USD' in total_value:
        return float(total_value.replace('$', '').replace('USD', '').strip())
    elif 'TL' in total_value:
        return float(total_value.replace('TL', '').replace(',', '.').strip()) / 27.0  # Adjust the exchange rate
    else:
        return 0

# Apply the function to clean and convert the 'Total' column
df['Total (USD)'] = df['Total'].apply(process_total)

# Group by Items and calculate purchases and refunds
summary = {}
for _, row in df.iterrows():
    item = row['Items']
    transaction_type = row['Type']
    total = row['Total (USD)']

    if item not in summary:
        summary[item] = {'Purchase': [], 'Refund': []}

    if transaction_type == 'Purchase':
        summary[item]['Purchase'].append(total)
    elif transaction_type == 'Refund':
        summary[item]['Refund'].append(total)

# Display the summary in the desired format
for item, transactions in summary.items():
    purchase_sum = " + ".join(f"{p:.2f} USD" for p in transactions['Purchase'])
    refund_sum = " - ".join(f"{r:.2f} USD" for r in transactions['Refund'])

    if purchase_sum or refund_sum:
        display_line = f"{item} ="
        if purchase_sum:
            display_line += f" +{purchase_sum}"
        if refund_sum:
            display_line += f" - {refund_sum}"
        print(display_line)
```
