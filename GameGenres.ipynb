```python
import pandas as pd

# Load the CSV file
data_url = "https://raw.githubusercontent.com/YavuzCanAtalay/DSA210_TermProject/refs/heads/main/DataFile/Games_Categories.csv"
df = pd.read_csv(data_url)

# Initialize counters for category occurrences and total categories per game
category_counts = {}
total_categories = 0

# Display all games with their categories, ensuring no duplicate categories and skipping Test Server/Public Test games
for index, row in df.iterrows():
    game_number = row['Number']
    game_name = row['Name']

    # Skip games with "Test Server" or "Public Test" in the name
    if "Test Server" in game_name or "Public Test" in game_name:
        continue

    # Collect unique categories
    categories = list(dict.fromkeys([str(row['Category1']), str(row['Category2']), str(row['Category3'])]))
    categories = [cat for cat in categories if cat != 'nan']  # Exclude any NaN categories

    # Update total categories count
    total_categories += len(categories)

    # Update category occurrence counts
    for category in categories:
        category_counts[category] = category_counts.get(category, 0) + 1

    # Display the game and its categories
    categories_display = ', '.join(categories)
    print(f"{game_number}. {game_name}: {categories_display}\n")


# Display total categories for all games
print(f"\nTotal number of categories across all games: {total_categories}\n")

# Display each category's occurrence, sorted from highest to lowest
print("Category occurrences (sorted):")
sorted_categories = sorted(category_counts.items(), key=lambda x: x[1], reverse=True)
for category, count in sorted_categories:
    print(f"{category}: {count}")

# Extract and display categories with less than 3 occurrences
print("\nSimplified version:")
rare_categories = [(category, count) for category, count in category_counts.items() if count>= 3]
for idx, (category, count) in enumerate(rare_categories, start=1):
    print(f"{idx}. {category}: {count}")
```
